```python
def priznak(p,t,n):
    
    t = np.array(t)
    w = np.arange(n, dtype=float)
    A = np.arange(n, dtype=float)
    
    for i in range(0,n):
        w[i] = t[i]/np.sum(t)
        
    for i in range(0,n):
        A[i] = p[i]*w[i]/(np.sum(p*w))    
    
    mean = np.mean(t)
    weighted_mean = np.sum(p*t)
    geometric_mean = gmean(t)
    weighted_geometric_mean = (t**p).prod()
    entropy_priznak = entropy(w)
    weighted_entropy = entropy(A)
    range_priznak = np.max(t) - np.min(t)
    weighted_range = p[t.argmax()]*np.max(t) - p[t.argmin()]*np.min(t)
    standard_deviation = np.std(t)
    
    weighted_stats = DescrStatsW(t, weights=p, ddof=0)
    weighted_standard_deviation = weighted_stats.std
    
    prznk = np.arange(10, dtype=float)
    prznk[0] = mean
    prznk[1] = weighted_mean
    prznk[2] = geometric_mean
    prznk[3] = weighted_geometric_mean
    prznk[4] = entropy_priznak
    prznk[5] = weighted_entropy
    prznk[6] = range_priznak
    prznk[7] = weighted_range
    prznk[8] = standard_deviation
    prznk[9] = weighted_standard_deviation
        
    return(prznk)    
```

```python
def material(p,t1,t2,t3,t4,t5,t6,t7,t8,n):
    data = sc_min.query("number_of_elements in [1]")
    data = data.drop('critical_temp', 1)
    data = data.drop('material', 1)
    data = data.iloc[0,:]
    
    data.iloc[0] = n
    for i in range(1,11):
        data.iloc[i] = priznak(p,t1,n)[i-1]
    for i in range(11,21):
        data.iloc[i] = priznak(p,t2,n)[i-11]
    for i in range(21,31):
        data.iloc[i] = priznak(p,t3,n)[i-21]
    for i in range(31,41):
        data.iloc[i] = priznak(p,t4,n)[i-31]
    for i in range(41,51):
        data.iloc[i] = priznak(p,t5,n)[i-41]
    for i in range(51,61):
        data.iloc[i] = priznak(p,t6,n)[i-51]
    for i in range(61,71):
        data.iloc[i] = priznak(p,t7,n)[i-61]
    for i in range(71,81):
        data.iloc[i] = priznak(p,t8,n)[i-71]
    data = data.to_frame().T
    return(data)
        
    
```
